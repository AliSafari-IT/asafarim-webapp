# ASAFARIM

## Frontend using multiple frameworks
On the front-end, it's generally best to stick to one framework across the whole user interface to ensure consistency and simplify state management. However, in our case, we tried to use multiple frameworks as micro-frontends, where different parts of our application are built and deployed as independent applications.

### React TS UI
`yarn create react-app ui-react --template typescript`

### Vue UI
`yarn create vue-app ui-vue --template typescript`  

### Angular UI
`yarn create angular-app ui-angular --template typescript`
